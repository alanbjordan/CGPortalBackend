# Set up openai

```python
# Openai Setup
from openai import OpenAI
import os

# Set up the OpenAI API key to interact with the GPT models
client = OpenAI(
    api_key=os.environ.get("OPENAI_API_KEY"),  # This is the default and can be omitted
)
if not client:
    raise ValueError("Please set the VA_AUTOMATION_API_KEY environment variable.")
```

```python
EMBEDDING_MODEL_SMALL = "text-embedding-3-small"
DATABASE_URL = os.environ.get("DATABASE_URL")

def get_embedding_small(text: str) -> list[float]:
    """
    Generate embeddings using OpenAI's 'text-embedding-ada-002' model.
    Returns the embedding vector (list of floats).
    """
    response = client.embeddings.create(
        model=EMBEDDING_MODEL_SMALL,
        input=text
    )
    return response
```

```python
embeddint_test = get_embedding_small("Hello, world!")
```
